```python
# Coordinates of nearby proximities (given / assumed in the lab)

# Coastline (nearest coastline to launch site)
coastline_lat = 28.56367
coastline_lon = -80.57163

# Railway
railway_lat = 28.572872
railway_lon = -80.585331

# Highway
highway_lat = 28.573255
highway_lon = -80.646895

# City (nearest city)
city_lat = 28.561857
city_lon = -80.577366
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians
from folium.features import DivIcon

# TASK 3: Calculate the distances between a launch site to its proximities

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# --- get launch site coords (use existing vars if notebook already set them) ---
if 'launch_site_lat' not in globals() or 'launch_site_lon' not in globals():
    launch_site_lat = float(spacex_df.iloc[0]['Lat'])
    launch_site_lon = float(spacex_df.iloc[0]['Long'])

launch_site_coordinate = [launch_site_lat, launch_site_lon]

# Helper to draw line + distance label to a target point
def add_distance_line(map_obj, start_coord, end_coord, label):
    distance_km = calculate_distance(start_coord[0], start_coord[1], end_coord[0], end_coord[1])

    # line
    folium.PolyLine(locations=[start_coord, end_coord], weight=2).add_to(map_obj)

    # marker + label at the end point
    folium.Marker(
        end_coord,
        icon=DivIcon(
            icon_size=(200, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color: #000;"><b>{label}: {distance_km:.2f} km</b></div>'
        )
    ).add_to(map_obj)

    return distance_km

# --- distances to proximities (these coords are typically provided earlier in the notebook) ---
# If you get NameError on any of these (coastline_lat, etc.), scroll up and run the cell that defines them.

# Coastline
coastline_distance = add_distance_line(
    site_map,
    launch_site_coordinate,
    [coastline_lat, coastline_lon],
    "Coastline"
)

# Railway
railway_distance = add_distance_line(
    site_map,
    launch_site_coordinate,
    [railway_lat, railway_lon],
    "Railway"
)

# Highway
highway_distance = add_distance_line(
    site_map,
    launch_site_coordinate,
    [highway_lat, highway_lon],
    "Highway"
)

# City
city_distance = add_distance_line(
    site_map,
    launch_site_coordinate,
    [city_lat, city_lon],
    "City"
)

site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Create markers with distance to closest city, railway, highway
# and draw lines between each marker and the launch site

def add_marker_and_line(map_obj, start_coord, end_coord, distance_km, color='#d35400'):
    # Line
    folium.PolyLine(
        locations=[start_coord, end_coord],
        weight=1,
        color=color
    ).add_to(map_obj)
    
    # Distance marker
    folium.Marker(
        end_coord,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:%s;"><b>%s</b></div>' 
                 % (color, "{:10.2f} KM".format(distance_km)),
        )
    ).add_to(map_obj)

# Coordinates
launch_site_coordinate = [launch_site_lat, launch_site_lon]

city_coordinate = [city_lat, city_lon]
railway_coordinate = [railway_lat, railway_lon]
highway_coordinate = [highway_lat, highway_lon]

# Distances
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# Add markers and lines
add_marker_and_line(site_map, launch_site_coordinate, city_coordinate, distance_city)
add_marker_and_line(site_map, launch_site_coordinate, railway_coordinate, distance_railway)
add_marker_and_line(site_map, launch_site_coordinate, highway_coordinate, distance_highway)

site_map
```
